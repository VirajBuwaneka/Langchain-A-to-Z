```python

from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI

# 1. Define your model
model = ChatOpenAI(model="gpt-3.5-turbo")

# 2. Create a prompt template
prompt = ChatPromptTemplate.from_template(
    "Write a POEM about {topic}"
)

# 3. Create a parser
parser = StrOutputParser()

# 4. Combine them into a chain
chain = prompt | model | parser

# 5. Run it
result = chain.invoke({"topic": "data science"})
print(result)
```

```python
question="My name is Viraj Buwaneka"
response=chain.invoke({"question":question})
print(response)
```

### Initialized the Prompt Template For Dynamic Interctions

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import SystemMessage, HumanMessage,AIMessage

#create a prompt template using MessagePlaceholder for the questions
prompt=ChatPromptTemplate.from_messages([
    SystemMessage(content="You are an intelligent chatbot.Answer the Following questions"),
    MessagesPlaceholder(variable_name="question"),
])
```

```python
#chain the prompt,LLM and Output parser

chain=prompt|llm|parser
```

```python
question="my name is Viraj Buwaneka"

response=chain.invoke({"question":[HumanMessage(content=question)]})
print(response)
```

```python
question="Who i am?"

response=chain.invoke({"question":[HumanMessage(content=question)]})
print(response)
```

### Initialize the Prompt Template to Handle Dynamic Conversation History

```python
#Create a prompt template with predefined conversational history

prompt=ChatPromptTemplate.from_messages([
    SystemMessage(content="You are an intelligent chatbot.Answer the Following questions"),
    MessagesPlaceholder(variable_name="history"),
    MessagesPlaceholder(variable_name="question")
])
```

```python
#Define conversation history
history=[
    HumanMessage(content="My name is Viraj Buwaneka"),
    AIMessage(content="Hello Viraj! How can I assist you today?"),
    HumanMessage(content="What is 2+2"),
    AIMessage(content="4")
]
```

```python
# Chain the prompt, LLM and Output parser
chain=prompt|llm|parser
```

```python
question="Who i Am?"
response=chain.invoke({"history":history,"question":[HumanMessage(content=question)]})
print(response)
```

### Update and Display Conversation History

```python
# Extend the history with the latest question and response
history.extend([
    HumanMessage(content=question),
    AIMessage(content=response)
])
```

```python
question="What is my last question"
response=chain.invoke({"history":history,"question":[HumanMessage(content=question)]})
history.extend([
    HumanMessage(content=question),
    AIMessage(content=response)
])

print(response)
```
