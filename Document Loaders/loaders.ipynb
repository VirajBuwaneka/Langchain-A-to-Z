### Load Text Files

```python
from langchain_community.document_loaders import TextLoader

# Initialize the text Loader with file path
loader = TextLoader('example_txt_file.txt')

txt_data = loader.load()

print(txt_data[0].page_content)
```

```python
txt_data[0].metadata
```

### Loading text file from a Directory

```python
!pip install unstructured -qU #package
```

```python
from langchain_community.document_loaders import DirectoryLoader

loader=DirectoryLoader("txt_folder",glob="*.txt")

dataset=loader.load()

for data in dataset:
    print("-----------------------")
    print(data.page_content)
    print(data.metadata)
    
```

### Loading PDF files

```python
!pip install pypdf -qU
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("example_pdf_file.pdf")

pdf_data=loader.load()

print(pdf_data[0].page_content)
print("-----------------------------")
print(pdf_data[0].metadata)
```

### Loading pdf file from a Directory

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

loader=PyPDFDirectoryLoader("pdf_folder")
pdf_dataset=loader.load()
for pdf in pdf_dataset:
    print("-----------------------")
    print(pdf.page_content)
    print(pdf.metadata)
```

### Loading CSV file

```python
from langchain_community.document_loaders import CSVLoader

loader=CSVLoader("example_csv_file.csv")

csv_data=loader.load()

print(csv_data[0].page_content)
print("-----------------------------")
print(csv_data[0].metadata)
```

### Loading HTML Data

```python
from langchain_community.document_loaders import BSHTMLLoader

loader=BSHTMLLoader("example_html_file.html")
html_data=loader.load()

print(html_data[0].page_content)
print("-----------------------------")  
print(html_data[0].metadata)
```
